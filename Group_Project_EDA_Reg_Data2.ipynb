```python
# -*- coding: utf-8 -*-
"""
Created on Tue Oct  5 10:52:31 2021

@author: ruchuan2
"""

## Group Project for the machine learning course

## Data 2

## read the MLF_GP2_EconCycle
import pandas as pd

data = pd.read_csv("C:/Users/ruchuan2/Box/IE 517 Machine Learning in FIN Lab/Group Project/MLF_GP2_EconCycle.csv")
del data["Date"]

## Explantory analysis (EDA)
data.head

## summary of the data

summary = data.describe()
print(summary)

## The boxplots of  T1Y index to T10Y index
import matplotlib.pyplot as plt

Treasury_bill = data.iloc[:,0:6]

plt.boxplot(Treasury_bill.values)
plt.xlabel("Maturity")
plt.ylabel("Yield")
plt.show()

## The distribution of each treasury bill is pretty similar.

## The boxplots of commerical paper yields

commerical_yields =  data.iloc[:,6:9]

plt.boxplot(commerical_yields.values)
plt.xlabel("Maturity")
plt.ylabel("Yield")
plt.show()

## The variation of USPHCI

index = data["USPHCI"]
plt.plot(index)
plt.xlabel("USPHCI Index")

## The heat map of pearson correlation coefficients of all treasuty bills
import seaborn as sns
import numpy as np

corr_matrix = np.corrcoef(Treasury_bill.values.T)
hm = sns.heatmap(corr_matrix,
                 cbar=True,
                 annot=True,
                 square=True,
                 yticklabels=['T1Y Index', 'T2Y Index', 'T3Y Index', 'T5Y Index', 'T7Y Index', 'T10Y Index'],
                 xticklabels=['T1Y Index', 'T2Y Index', 'T3Y Index', 'T5Y Index', 'T7Y Index', 'T10Y Index'])
plt.show()

## The heat map of pearson correlation coefficients of all commerical paper yields

corr_matrix = np.corrcoef(commerical_yields.values.T)
hm = sns.heatmap(corr_matrix,
                 cbar=True,
                 annot=True,
                 square=True,
                 yticklabels=['CP1M', 'CP3M', 'CP6M'],
                 xticklabels=['CP1M', 'CP3M', 'CP6M'])
plt.show()

## Regression model
################################ Predict the PCT 3MO FWD ####################################
## This time, I put T1Y Index to T5Y Index, and CP1M_T1Y to CP6M_T1Y, all as features

## standardlize

from sklearn.preprocessing import StandardScaler
from pandas import DataFrame

del data["CP1M"]
del data["CP3M"]
del data["CP6M"]
del data["USPHCI"]

sc = StandardScaler()
data = sc.fit_transform(data)
data = DataFrame(data)

X, y = data.iloc[:, 0:9], data.iloc[:, 9]

# Split the dataset into a training and a testing set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
print( X_train.shape, y_train.shape, X_test.shape, y_test.shape)

## Linear Regression Model

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
y_train_pred_linear = lr.predict(X_train)
y_test_pred_linear = lr.predict(X_test)

### RMSE
import math

from sklearn.metrics import mean_squared_error
rmse_train_linear = math.sqrt(mean_squared_error(y_train, y_train_pred_linear))
rmse_test_linear = math.sqrt(mean_squared_error(y_test, y_test_pred_linear))

## calculate R-squre

from sklearn.metrics import r2_score
r_square_train_linear = r2_score(y_train, y_train_pred_linear)
r_square_test_linear = r2_score(y_test, y_test_pred_linear)

######### SVR

from sklearn import svm
svr = svm.SVR(kernel='rbf')
svr.fit(X_train,y_train)
y_train_pred_svm = svr.predict(X_train)
y_test_pred_svm = svr.predict(X_test)

### RMSE

from sklearn.metrics import mean_squared_error
rmse_train_svm = math.sqrt(mean_squared_error(y_train, y_train_pred_svm))
rmse_test_svm = math.sqrt(mean_squared_error(y_test, y_test_pred_svm))

## calculate R-squre

from sklearn.metrics import r2_score
r_square_train_svm = r2_score(y_train, y_train_pred_svm)
r_square_test_svm = r2_score(y_test, y_test_pred_svm)

##########################  PCT 6MO FWD #############################

X, y = data.iloc[:, 0:9], data.iloc[:, 10]

# Split the dataset into a training and a testing set

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
print( X_train.shape, y_train.shape, X_test.shape, y_test.shape)

## Linear Regression Model

lr = LinearRegression()
lr.fit(X_train,y_train)
y_train_pred_linear = lr.predict(X_train)
y_test_pred_linear = lr.predict(X_test)

### RMSE

rmse_train_linear = math.sqrt(mean_squared_error(y_train, y_train_pred_linear))
rmse_test_linear = math.sqrt(mean_squared_error(y_test, y_test_pred_linear))

## calculate R-squre

r_square_train_linear = r2_score(y_train, y_train_pred_linear)
r_square_test_linear = r2_score(y_test, y_test_pred_linear)

###### SVR

svr = svm.SVR(kernel='rbf')
svr.fit(X_train,y_train)
y_train_pred_svm = svr.predict(X_train)
y_test_pred_svm = svr.predict(X_test)

### RMSE

rmse_train_svm = math.sqrt(mean_squared_error(y_train, y_train_pred_svm))
rmse_test_svm = math.sqrt(mean_squared_error(y_test, y_test_pred_svm))

## calculate R-squre

r_square_train_svm = r2_score(y_train, y_train_pred_svm)
r_square_test_svm = r2_score(y_test, y_test_pred_svm)

##########################  PCT 6MO FWD #############################

X, y = data.iloc[:, 0:9], data.iloc[:, 11]

# Split the dataset into a training and a testing set

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
print( X_train.shape, y_train.shape, X_test.shape, y_test.shape)

## Linear Regression Model

lr = LinearRegression()
lr.fit(X_train,y_train)
y_train_pred_linear = lr.predict(X_train)
y_test_pred_linear = lr.predict(X_test)

### RMSE

rmse_train_linear = math.sqrt(mean_squared_error(y_train, y_train_pred_linear))
rmse_test_linear = math.sqrt(mean_squared_error(y_test, y_test_pred_linear))

## calculate R-squre

r_square_train_linear = r2_score(y_train, y_train_pred_linear)
r_square_test_linear = r2_score(y_test, y_test_pred_linear)

### SVR

svr = svm.SVR(kernel='rbf')
svr.fit(X_train,y_train)
y_train_pred_svm = svr.predict(X_train)
y_test_pred_svm = svr.predict(X_test)

### RMSE

rmse_train_svm = math.sqrt(mean_squared_error(y_train, y_train_pred_svm))
rmse_test_svm = math.sqrt(mean_squared_error(y_test, y_test_pred_svm))

## calculate R-squre

r_square_train_svm = r2_score(y_train, y_train_pred_svm)
r_square_test_svm = r2_score(y_test, y_test_pred_svm)
```
